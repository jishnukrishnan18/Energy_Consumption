```python
#1.Energy Consumption Trend (Time Series)
plt.figure(figsize=(20,8))
plt.plot(df['Date'], df['Daily_Consumption'])
plt.title("Energy Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Consumption")
plt.show()
```

```python
#2.What is the relationship between Max Temperature and Energy Consumption?
plt.scatter(df['Max_Temp'], df['Daily_Consumption'])
plt.xlabel("Max Temperature")
plt.ylabel("Energy Consumption")
plt.title("Max Temp vs Energy")
plt.show()
```

```python
#3.How does rainfall affect energy consumption
plt.scatter(df['Rainfall'], df['Daily_Consumption'])
plt.xlabel("Rainfall")
plt.ylabel("Energy Consumption")
plt.title("Rainfall vs Energy")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # make sure it's datetime

df['Month'] = df['Date'].dt.month
df
```

```python
#4.How does energy consumption vary across different months?
Monthly= df.groupby('Month')['Daily_Consumption'].mean()
Monthly.plot(kind='bar')
plt.title("Average Monthly Consumption")
plt.xlabel("Month")
plt.ylabel("Consumption")
plt.show()
```

```python
#5.What is the relationship between weather variables and energy consumption
corr = df[['Wind_Speed','Rainfall','Max_Temp','Min_Temp','Daily_Consumption']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#6.Which is more influential: Max_Temp or Min_Temp
plt.scatter(df['Max_Temp'], df['Daily_Consumption'], label='Max_Temp')
plt.scatter(df['Min_Temp'], df['Daily_Consumption'], label='Min_Temp')
plt.legend()
plt.show()
```

```python
#7.What could be the reason for peak consumption days
peak = df[df['Daily_Consumption'] > df['Daily_Consumption'].quantile(0.90)]
peak_means = peak[['Max_Temp','Min_Temp','Rainfall','Wind_Speed']].mean()
peak_means.plot(kind='bar')
plt.title("Weather Conditions During Peak Consumption Days")
plt.ylabel("Average Values")
plt.show()
```

```python
#8.which day has the high energy consumption
df['Day'] = df['Date'].dt.day_name()

daywise = df.groupby('Day')['Daily_Consumption'].mean()

daywise = daywise.reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
])

daywise.plot(kind='bar')
plt.title("Day-wise Energy Consumption")
plt.show()
```

```python
x=df[['Wind_Speed','Rainfall','Min_Temp','Daily_Consumption','Month','Month_Name','Day', 'Year']].values
y=df['Max_Temp'].values
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
x = df[['Wind_Speed','Rainfall','Min_Temp','Daily_Consumption',
        'Month','Month_Name','Day','Year']]
y = df['Max_Temp']
categorical_cols = ['Month_Name', 'Day']
ct = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
], remainder='passthrough')
x = ct.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=99)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
from sklearn import metrics
print('Root mean sqaured error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("r2 score:",score*100,'%')
```

## CONCLUSION

After performing Exploratory Data Analysis (EDA), Regression modeling, and Backward Elimination, we can summarize the key findings:

### 1.Data understand & Cleaning

Data cleaning was performed to improve the quality and reliability of the dataset. Column names were standardized, dataset structure and statistics were checked, missing values were handled, duplicate records were removed, and the date format was corrected. These steps ensured the data became clean, consistent, and ready for analysis and machine learning modeling.

### 2. Exploratory Data Analysis (EDA)
- Energy consumption changed over time
- Temperature strongly affected usage
- Rainfall had some impact
- Consumption varied by month and year
- Peak days showed highest energy demand
- EDA identified key consumption patterns

### 3. Linear Regression
- Built a Multiple Linear Regression model to predict energy consumption.
- Features used:
  - Date,Wind_Speed,Rainfall,Max_Temp,Min_Temp,
        Month,Day,Month_Name,Year.
- Model showed:
  - Strong predictive relationship.
  - Good performance based on R² and RMSE.

### 4. Polynomial Regression 
- Polynomial Regression was applied to capture non-linear patterns.
- However, the dataset showed a mostly linear relationship between variables (especially Max_Temp vs Min_Temp).
- Key Finding:
  - Polynomial Regression did NOT significantly improve the model.
- Reason:
  - Data is already linear.
  - Adding polynomial term (x²) does not add useful information.

### 5. Backward Elimination
- Backward Elimination was applied to identify insignificant features in the model
- P-values from the OLS summary were checked for feature significance
- No feature had a high insignificant p-value requiring removal
- All selected variables contributed meaningfully to the prediction
- Therefore, no columns were removed from the dataset
- The final model retained all features for better prediction performance.

### Final Conclusion
- Data cleaning and EDA identified key energy consumption patterns and important influencing factors like temperature and season.
- Linear Regression predicted energy consumption effectively, while Polynomial Regression showed no major improvement.
- Backward Elimination confirmed all features were significant, so no columns were removed.
